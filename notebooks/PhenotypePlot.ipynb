```python
# import packages
import os, sys, glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# set the path to the data``
sys.path.append("../")
```

```python
# load data
phenotypes = pd.read_csv("../pennycress_pods/test/test_measurements/measurements.csv").iloc[:, 1:]
phenotypes.head()
```

```python
# plot seed count v. seed area
sns.lmplot(x="seed_count", y="seed area", data=phenotypes, ci=None, line_kws={'color': 'red'})
plt.title("Seed Count v. Seed Area")
```

```python
# plot seed count v. seed area
sns.lmplot(x="wing area", y="seed_count", data=phenotypes, ci=None, line_kws={'color': 'red'})
plt.title("Wing Area v. Seed Count")
```

```python
# plot seed count v. seed area
sns.lmplot(x="env area", y="seed_count", data=phenotypes, ci=None, line_kws={'color': 'red'})
plt.title("Env Area v. Seed Count")
```

```python
# plot seed count v. seed area
sns.lmplot(x="env area", y="seed area", data=phenotypes, ci=None, line_kws={'color': 'red'})
plt.title("Env Area v. Seed Area")
```
